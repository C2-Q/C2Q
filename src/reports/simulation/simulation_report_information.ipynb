The qaoa_dict_saved and vqe_dict_saved dictionaries and the counts dictionaries can be loaded with this script.

qaoa_dict_saved and vqe_dict_saved dictionaries contain the parameters of the circuits and the expectation values of every iteration. counts dictionaries contain the sampled counts of the runs.

```python
import pickle
```

```python
with open('data/qaoa_dict_saved.pkl', 'rb') as f:
    qaoa_dict_saved = pickle.load(f)
```

```python
qaoa_dict_saved
```

```python
with open('data/qaoa_counts.pkl', 'rb') as f:
    qaoa_counts = pickle.load(f)
```

```python
qaoa_counts
```
